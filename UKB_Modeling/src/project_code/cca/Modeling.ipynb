```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Your specific column names
cholangitis_columns = ["Cholangitis"]

cirrhosis_columns = [
    "Alcoholic cirrhosis",
    "Hepatic fibrosis",
    "Hepatic sclerosis",
    "Hepatic fibrosis/sclerosis",
    "Primary biliary cirrhosis",
    "Secondary biliary cirrhosis",
    "Biliary cirrhosis",
    "Liver cirrhosis",
    "Complications of cirrhosis"
]

# Your comprehensive list of PAR comorbidities (including cirrhosis and cholangitis)
all_par_comorbidities = [
    "Cholangitis",
    "Alcoholic cirrhosis",
    "Hepatic fibrosis",
    "Hepatic sclerosis",
    "Hepatic fibrosis/sclerosis",
    "Primary biliary cirrhosis",
    "Secondary biliary cirrhosis",
    "Biliary cirrhosis",
    "Liver cirrhosis",
    "Complications of cirrhosis",
    "Reactive hepatitis",
    "Granulomatous hepatitis",
    "Autoimmune hepatitis",
    "Alcoholic fatty liver",
    "Alcoholic hepatitis",
    "Alcoholic fibrosis",
    "Alcoholic hepatic failure",
    "Alcoholic liver disease",
    "NASH",
    "Inflammatory liver disease",
    "NAFLD",
    "AATD",
    "Hemochromatosis",
    "Alcoholism",
    "Chronic Hepatitis B",
    "Chronic Hepatitis C",
    "Obstruction of bile duct",
    "Perforation of bile duct",
    "Fistula of bile duct",
    "Spasm of sphincter of Oddi",
    "Biliary Cyst",
    "Other specified diseases of biliary tract",
    "Disease of biliary tract, unspecified",
    "Calculus of gallbladder with acute cholecystitis",
    "Calculus of gallbladder with other cholecystitis",
    "Calculus of gallbladder without cholecystitis",
    "Calculus of bile duct with cholangitis",
    "Calculus of bile duct with cholecystitis",
    "Calculus of bile duct without cholangitis or cholecystitis",
    "Other cholelithiasis",
    "Elevated_AST",
    "Elevated_ALT",
    "Elevated_GGT",
    "Elevated_Liver_Enzymes",
    "Elevated_AP",
    "Alcohol-induced chronic pancreatitis",
    "Other chronic pancreatitis",
    "Sepsis",
    "IBD",
    "Chronic kidney disease",
    "Thrombopenia",
    "Anemia",
    "Yellow Fever",
    "Schistosomiasis",
    "Dengue Fever"
]

def categorize_double_donut_patients(df, cholangitis_cols, cirrhosis_cols, all_par_cols):
    """
    Create categorization for double donut visualization
    """
    # Filter for CCA patients only (status = 1)
    cca_patients = df[df['status'] == 1].copy()

    # Create aggregate diagnosis indicators
    cca_patients['has_cirrhosis'] = cca_patients[cirrhosis_cols].max(axis=1)
    cca_patients['has_cholangitis'] = cca_patients[cholangitis_cols].max(axis=1)

    # Check which PAR columns actually exist in the dataframe
    existing_par_cols = [col for col in all_par_cols if col in df.columns]
    if existing_par_cols:
        cca_patients['is_par'] = cca_patients[existing_par_cols].max(axis=1)
    else:
        cca_patients['is_par'] = 0

    def assign_inner_category(row):
        """Inner donut categories"""
        cholangitis = row['has_cholangitis']
        cirrhosis = row['has_cirrhosis']
        is_par = row['is_par']

        if cholangitis == 1 and cirrhosis == 1:
            return 'Screened (Both)'
        elif cirrhosis == 1 and cholangitis == 0:
            return 'Screened (Cirrhosis)'
        elif cholangitis == 1 and cirrhosis == 0:
            return 'Screened (Cholangitis)'
        elif is_par == 1:
            return 'Other comorbidities'
        else:
            return 'No risk factors'

    def assign_outer_category(row):
        """Outer donut categories"""
        cholangitis = row['has_cholangitis']
        cirrhosis = row['has_cirrhosis']
        is_par = row['is_par']

        # PAR includes cholangitis, cirrhosis, or other comorbidities
        if cholangitis == 1 or cirrhosis == 1 or is_par == 1:
            return 'PAR'
        else:
            return 'Unscreened'

    cca_patients['inner_category'] = cca_patients.apply(assign_inner_category, axis=1)
    cca_patients['outer_category'] = cca_patients.apply(assign_outer_category, axis=1)

    return cca_patients

def create_double_donut_plot(cca_patients):
    """
    Create triple-layered donut plot with properly aligned labels
    Inner ring: detailed categories
    Middle ring: PAR classification
    Outer ring: Screened/Unscreened classification
    """
    # Get counts for inner layer
    inner_counts = cca_patients['inner_category'].value_counts()

    # Define order and colors for inner ring
    inner_order = ['No risk factors', 'Other comorbidities', 'Screened (Cirrhosis)',
                   'Screened (Cholangitis)', 'Screened (Both)']
    inner_colors = {
        'No risk factors': "#430154CE" ,        # Dark Green
        'Other comorbidities': "#414487",    # Light Green
        'Screened (Cirrhosis)': "#2A788E",   # Tomato
        'Screened (Cholangitis)': "#22A884", # Orange
        'Screened (Both)': "#5AC73C"         # Dark Red
    }

    # Prepare data for inner ring (only non-zero categories)
    inner_data = []
    inner_labels = []
    inner_color_list = []

    for category in inner_order:
        if category in inner_counts and inner_counts[category] > 0:
            inner_data.append(inner_counts[category])
            inner_labels.append(category)
            inner_color_list.append(inner_colors[category])

    # Prepare data for PAR ring (middle layer)
    par_data = []
    par_labels = []
    par_colors = []

    for category in inner_order:
        if category in inner_counts and inner_counts[category] > 0:
            count = inner_counts[category]
            par_data.append(count)

            if category in ['Other comorbidities', 'Screened (Cirrhosis)', 'Screened (Cholangitis)', 'Screened (Both)']:
                par_labels.append('PAR')
                par_colors.append("#5AC73C")  # Light green for PAR
            else:
                par_labels.append('')  # No label for non-PAR
                par_colors.append('white')  # White for non-PAR

    # Prepare data for Screened ring (outer layer) - FIXED COLOR ASSIGNMENT
    screened_data = []
    screened_labels = []
    screened_colors = []

    for category in inner_order:
        if category in inner_counts and inner_counts[category] > 0:
            count = inner_counts[category]
            screened_data.append(count)

            if category in ['Screened (Cirrhosis)', 'Screened (Cholangitis)', 'Screened (Both)']:
                screened_labels.append('')
                screened_colors.append("white")  # White for Screened (no color)
            else:

                screened_colors.append('#430154CE')  # Purple color for Unscreened
                if category == 'No risk factors':
                    screened_labels.append('Unscreened')

    # Create the plot
    fig, ax = plt.subplots(figsize=(12, 10))

    # Create screened ring (outermost)
    screened_wedges, screened_texts, screened_autotexts = ax.pie(
        screened_data,
        radius=0.82,
        colors=screened_colors,
        autopct='',  # No percentages on outer rings
        startangle=90,
        wedgeprops=dict(width=0.05, edgecolor="#2C0154", linewidth=0)
    )

    # Create PAR ring (middle)
    par_wedges, par_texts, par_autotexts = ax.pie(
        par_data,
        radius=0.75,
        colors=par_colors,
        autopct='',  # No percentages on outer rings
        startangle=90,
        wedgeprops=dict(width=0.05, edgecolor="#D2FD25", linewidth=0)
    )

    # Create inner donut (innermost) - store wedge information
    inner_wedges, inner_texts, inner_autotexts = ax.pie(
        inner_data,
        radius=0.6,
        colors=inner_color_list,
        autopct=lambda pct: f'{pct:.1f}%\n({int(pct/100 * sum(inner_data))})',
        startangle=90,
        pctdistance=0.8,
        wedgeprops=dict(width=0.4, edgecolor='white', linewidth=2)
    )

    # Remove default labels
    for text in screened_texts + par_texts + inner_texts:
        text.set_text('')

    # Style the percentage text conditionally
    for autotext, wedge, label in zip(inner_autotexts, inner_wedges, inner_labels):
        if label in ["No risk factors", "Other comorbidities"]:
            autotext.set_color("white")
        else:
            autotext.set_color("black")
        autotext.set_fontsize(11)
        autotext.set_weight("normal")

    # Add title
    ax.set_title('CCA cohorts', fontsize=24, fontweight='normal', pad=0, horizontalalignment='center', verticalalignment='center', y=0.9, x=0.3)

    # Add center text
    total_patients = len(cca_patients)
    ax.text(0, 0, f'All CCA\n{total_patients}',
            horizontalalignment='center', verticalalignment='center',
            fontsize=20, fontweight='normal')

    # Add labels for inner ring using wedge information
    for i, (wedge, label) in enumerate(zip(inner_wedges, inner_labels)):
        # Get the middle angle of the wedge
        angle = (wedge.theta1 + wedge.theta2) / 2
        # Convert to radians
        angle_rad = np.radians(angle)

        # Position label outside the inner ring (closer to plot)
        inner_radius = 0.6  # Reduced from 1.1 to bring labels closer
        x = inner_radius * np.cos(angle_rad)
        y = inner_radius * np.sin(angle_rad)

        # Simplify labels for display
        display_label = {
            'No risk factors': 'No Risk\nFactors',
            'Other comorbidities': 'Other\nComorbidities',
            'Screened (Cirrhosis)': 'Cirrhosis',
            'Screened (Cholangitis)': 'Cholangitis',
            'Screened (Both)': 'Both'
        }

        # Determine text alignment based on position
        ha = 'center'
        if x < -0.1:
            ha = 'right'
        elif x > 0.1:
            ha = 'left'

        ax.text(x, y, display_label.get(label, label),
               horizontalalignment=ha, verticalalignment='center',
               fontsize=15, fontweight='normal',
               bbox=dict(boxstyle="round,pad=0.3", facecolor='white', alpha=0.9))

    # Add labels for PAR ring (single label for the entire PAR ring)
    # Calculate the total angle span for PAR segments
    par_segments = [(wedge, label) for wedge, label in zip(par_wedges, par_labels) if label == 'PAR']

    if par_segments:
        # Find the middle of all PAR segments combined
        all_par_angles = []
        for wedge, _ in par_segments:
            # Add all angles within this wedge
            start_angle = wedge.theta1
            end_angle = wedge.theta2
            mid_angle = (start_angle + end_angle) / 2
            all_par_angles.append(mid_angle)

        # Use the middle angle of all PAR segments
        avg_angle = np.mean(all_par_angles)
        angle_rad = np.radians(avg_angle)

        # Position single PAR label
        par_radius = 0.97
        x = par_radius * np.cos(angle_rad)
        y = par_radius * np.sin(angle_rad)

        # Determine text alignment based on position
        ha = 'center'
        if x < -0.1:
            ha = 'right'
        elif x > 0.1:
            ha = 'left'

        ax.text(x, y, 'PAR',
               horizontalalignment=ha, verticalalignment='center',
               fontsize=18, fontweight='normal', style='italic',
               bbox=dict(boxstyle="round,pad=0.4", facecolor='#5AC74C', alpha=0.5))

    # --- Add label for UNSCREENED (single combined label like PAR) ---
    unscreened_segments = [(wedge, label) for wedge, label in zip(screened_wedges, screened_labels) if label == 'Unscreened']

    if unscreened_segments:
        all_unscreened_angles = []
        for wedge, _ in unscreened_segments:
            start_angle = wedge.theta1
            end_angle = wedge.theta2
            mid_angle = (start_angle + end_angle) / 2
            all_unscreened_angles.append(mid_angle)

        avg_angle = np.mean(all_unscreened_angles)
        angle_rad = np.radians(avg_angle)

        # Position label outside Screened ring
        unscreened_radius = 0.97
        x = unscreened_radius * np.cos(angle_rad)
        y = unscreened_radius * np.sin(angle_rad)

        ha = 'center'
        if x < -0.1:
            ha = 'right'
        elif x > 0.1:
            ha = 'left'

        ax.text(x, y, 'Unscreened',
                horizontalalignment=ha, verticalalignment='center',
                fontsize=18, fontweight='normal', style='italic',
                bbox=dict(boxstyle="round,pad=0.4", facecolor='#2C0154', alpha=0.5))

    # Set equal aspect ratio and remove axes
    ax.set_aspect('equal')
    ax.axis('off')

    plt.tight_layout()
    return fig, ax

def print_double_donut_summary(cca_patients):
    """
    Print summary matching the double donut structure
    """
    print("Double Donut Summary:")
    print("=" * 50)

    total = len(cca_patients)
    inner_counts = cca_patients['inner_category'].value_counts()
    outer_counts = cca_patients['outer_category'].value_counts()

    print("INNER RING (Risk Detail):")
    for category in ['No risk factors', 'Other comorbidities', 'Screened (Cirrhosis)',
                     'Screened (Cholangitis)', 'Screened (Both)']:
        count = inner_counts.get(category, 0)
        if count > 0:
            print(f"  {category}: {count} ({count/total*100:.1f}%)")

    print("\nOUTER RING (Main Categories):")
    for category in ['Unscreened', 'PAR']:
        count = outer_counts.get(category, 0)
        if count > 0:
            print(f"  {category}: {count} ({count/total*100:.1f}%)")

    print(f"\nTotal: {total} patients")

# Example usage:
cca_patients = categorize_double_donut_patients(df_all, cholangitis_columns, cirrhosis_columns, all_par_comorbidities)
print_double_donut_summary(cca_patients)
fig, ax = create_double_donut_plot(cca_patients)


out_dir = "C:/Users/felix/OneDrive - Uniklinik RWTH Aachen/CRC-1382-A11 - drive/public/projects/cca/visuals/"
plt.savefig((out_dir + "double_donut_cca_patients.svg"), dpi=300,transparent=True)
plt.show()
```
